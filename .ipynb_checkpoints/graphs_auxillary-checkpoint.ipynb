```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("person.csv",sep=";")
```

```python
df
```

```python
male = df[df["sex"] == "Male"]
male
```

```python
female = df[df["sex"] == "Female"]
female
```

```python
sizes = [len(male),len(female)]
sizes
```

```python
labels = ["Male","Female"]
labels
```

```python
fig, ax = plt.subplots()
ax.pie(sizes, labels=labels,autopct='%1.1f%%')
```
