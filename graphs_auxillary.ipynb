```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("person.csv",sep=";")
```

```python
df
```

```python
male = df[df["sex"] == "Male"]
male
```

```python
female = df[df["sex"] == "Female"]
female
```

```python
sizes = [len(male),len(female)]
sizes
```

```python
labels = ["Male","Female"]
labels
```

```python
fig, ax = plt.subplots()
ax.pie(sizes, labels=labels,autopct='%1.1f%%')
plt.savefig("gender_distribution.png")
```

```python
df.columns
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['weight'], bins=20, kde=True, color='skyblue', edgecolor='black')
plt.title('Weight Distribution', fontsize=16)
plt.xlabel('Weight (kg)', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.savefig("weight_distribution.png")  # Saves as a PNG file
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='height', y='weight', hue='sex', palette='Set2', alpha=0.7)
plt.title('Height vs. Weight', fontsize=16)
plt.xlabel('Height (cm)', fontsize=14)
plt.ylabel('Weight (kg)', fontsize=14)
plt.grid(alpha=0.3)
plt.legend(title='Gender', loc='upper left')
plt.savefig("height_vs_weight.png")  # Saves as a PNG file
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['height'], bins=20, kde=True, color='skyblue', edgecolor='black')
plt.title('Height Distribution', fontsize=16)
plt.xlabel('Height', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.savefig("height_distribution.png")  # Saves as a PNG file
plt.show()
```
